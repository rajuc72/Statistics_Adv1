####2. What are the types of random variables?

Discrete Random Variables: Take on countable values (e.g., number of heads in coin tosses, no.of cars , no.of mobile phones).
Continuous Random Variables: Take on an infinite number of values within a range (e.g., height, weight).

####3. What is the difference between discrete and continuous distributions?

Ans: Discrete Distributions: Deal with countable outcomes.
Continuous Distributions: Deal with outcomes that can take any value within a range.

####4.What are probability distribution functions (PDF)?

Ans: Probability Density Function (PDF) is used for continuous random variables. It provides the density of probabilities rather than the exact probabilities because the probability of a continuous variable taking any exact value is essentially zero.
Example: Imagine you have a uniformly distributed continuous random variable that represents the height of people in a certain town, ranging from 150 cm to 200 cm. The PDF in this case might look something like a rectangle where each height between 150 cm and 200 cm is equally likely.
If you wanted to find the probability that a person's height is between 160 cm and 170 cm, you would calculate the area under the PDF curve from 160 to 170.

####5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

Ans: Cumulative Distribution Function (CDF) represents the cumulative probability that a random variable will take a value less than or equal to a specific value. It‚Äôs the integral of the PDF for continuous variables or the sum of the probabilities for discrete variables.
Example: Using the same height example, the CDF would start at 0 at 150 cm, and increase linearly to 1 at 200 cm. If you want to know the probability that a person's height is less than or equal to 170 cm, you would look at the CDF value at 170 cm.
If the CDF at 170 cm is 0.4, it means there is a 40% chance that a randomly selected person‚Äôs height is less than or equal to 170 cm.
PDF: Represents probability of a specific value.
CDF: Represents cumulative probability up to a certain value.

####13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans: The Central Limit Theorem (CLT) is a fundamental principle in probability and statistics. It states that, given a sufficiently large sample size, the distribution of the sample means will approximate a normal distribution (bell curve), regardless of the shape of the population distribution.
Key Points:
Sample Size: The larger the sample size, the closer the sample mean distribution will be to a normal distribution.
Population Distribution: The original population distribution can be skewed, uniform, or any other.
Mean and Standard Deviation: The mean of the sample means will equal the population mean. The standard deviation of the sample means (known as the standard error) will be the population standard deviation divided by the square root of the sample size.
Example: Imagine you're measuring the heights of people in a town. If you take multiple random samples of a large enough size (e.g., 30 or more), the means of these samples will form a normal distribution, even if the original population distribution of heights isn't perfectly normal.

####14.How does the Central Limit Theorem relate to the normal distribution?

Ans: The CLT is crucial because it explains why the normal distribution (or bell curve) appears so frequently in statistics. Due to the CLT, we can make inferences about the population even when we don't know the exact distribution of the population.

Key Relationship:
Estimation: The normal distribution allows us to make probability statements about the sample means. For example, we can use Z-scores to find probabilities and confidence intervals.
Practical Application: Because of the CLT, many statistical methods and hypothesis tests assume a normal distribution. This makes the analysis robust and widely applicable.

####17. What are point estimates and interval estimates in statistics?

Ans: Point Estimate: A point estimate is a single value given as the estimate of a parameter of a population. It‚Äôs a specific numerical value.
Example: Estimating the average height of students in a class by measuring the height of a randomly selected student. If the measured height is 160 cm, this is your point estimate for the average height.
Interval Estimate: An interval estimate gives a range within which the parameter is expected to lie, providing more information about the variability.
Example: Instead of just saying the average height is 160 cm, you might say the average height is between 155 cm and 165 cm with 95% confidence. This range is called a confidence interval.

####18. What is the significance of confidence intervals in statistical analysis?

Ans: Confidence Intervals: A confidence interval is a range of values, derived from the sample data, that is likely to contain the value of an unknown population parameter.
Example: Suppose you survey a group of people to estimate the average number of hours they sleep. If you find the average to be 7 hours with a 95% confidence interval of [6.5, 7.5], you are 95% confident that the true average sleep time for the population is between 6.5 and 7.5 hours.
Significance: Confidence intervals provide a measure of uncertainty. They tell you how confident you can be in your estimate and give a range where the true parameter is likely to lie.

####9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
def calculate_and_plot_z_scores(data):
    # Calculate the mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data)
    # Calculate the Z-scores
    z_scores = (data - mean) / std_dev
    # Plot the histogram of Z-scores
    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.show()
    return z_scores

data = np.random.normal(50, 10, 1000)  # Generating random data from a normal distribution
z_scores = calculate_and_plot_z_scores(data)
```

####18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal Distribution


```python
from scipy.stats import norm
def calculate_z_score(data_point, mean, std_dev):
    return (data_point - mean) / std_dev
def plot_standard_normal_distribution(z_score):
    x = np.linspace(-4, 4, 1000)
    # Calculate the standard normal distribution (mean=0, std=1)
    pdf = norm.pdf(x, 0, 1)
    # Plot the standard normal distribution
    plt.plot(x, pdf, color='blue', label='Standard Normal Distribution')
    # Plot the Z-score on the standard normal distribution
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.fill_between(x, 0, pdf, where=(x <= z_score), color='red', alpha=0.3)

    plt.title('Standard Normal Distribution with Z-score')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Sample data point, mean, and standard deviation
data_point = 75
mean = 60
std_dev = 10
# Calculate the Z-score
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point}: {z_score:.2f}")
# Plot the standard normal distribution and highlight the Z-score
plot_standard_normal_distribution(z_score)
```

####19.Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import scipy.stats as stats
# Sample data
sample_data = [52, 55, 53, 54, 56, 51, 54, 53, 52, 55]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
# Population parameters
population_mean = 50
population_std_dev = 5
# Calculate the Z-score
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
print(f"Z-score: {z_score:.2f}")
# Determine the p-value for the Z-score
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
print(f"P-value: {p_value:.4f}")
# Determine the critical value for a 5% significance level (two-tailed test)
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)
print(f"Critical Z-value: ¬±{z_critical:.2f}")
# Compare the Z-score with the critical value
if abs(z_score) > z_critical:
    print("Reject the null hypothesis (H0). The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0). The sample mean is not significantly different from the population mean.")
```

####20.Create a confidence interval for a dataset using Python and interpret the result.

```python
sample_data = [52, 55, 53, 54, 56, 51, 54, 53, 52, 55]
sample_mean = np.mean(sample_data)
sample_std_dev = np.std(sample_data, ddof=1)
sample_size = len(sample_data)
# Define the confidence level
confidence_level = 0.95
alpha = 1 - confidence_level
# Calculate the critical Z-value (two-tailed)
z_critical = stats.norm.ppf(1 - alpha / 2)
# Calculate the standard error (SE)
standard_error = sample_std_dev / np.sqrt(sample_size)
# Calculate the margin of error (ME)
margin_of_error = z_critical * standard_error
# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Standard Deviation: {sample_std_dev:.2f}")
print(f"Sample Size: {sample_size}")
print(f"Critical Z-value: {z_critical:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: {confidence_interval}")
# Interpret the result
print(f"We are {confidence_level * 100}% confident that the true population mean lies within the range {confidence_interval}.")
```

```python
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)
# Calculate the mean and standard deviation of the original dataset
mean = np.mean(data)
std_dev = np.std(data)
# Apply Z-score normalization
normalized_data = (data - mean) / std_dev
# Calculate the mean and standard deviation of the normalized dataset
normalized_mean = np.mean(normalized_data)
normalized_std_dev = np.std(normalized_data)
# Print the means and standard deviations
print(f"Original Dataset Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}")
print(f"Normalized Dataset Mean: {normalized_mean:.2f}, Standard Deviation: {normalized_std_dev:.2f}")
# Visualize the original and normalized datasets
plt.figure(figsize=(12, 6))
# Histogram of the original dataset
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Original Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')
# Histogram of the normalized dataset
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Normalized Dataset')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

####28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

```python
from scipy.stats import skew, kurtosis
def calculate_skewness_kurtosis(data):
    # Calculate skewness
    skewness = skew(data)
    # Calculate kurtosis
    kurt = kurtosis(data)
    return skewness, kurt
# Generate a sample dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)
# Calculate skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)
# Print the results
print(f"Skewness of the dataset: {skewness:.2f}")
print(f"Kurtosis of the dataset: {kurt:.2f}")
# Interpretation
if skewness > 0:
    skewness_description = "positively skewed"
elif skewness < 0:
    skewness_description = "negatively skewed"
else:
    skewness_description = "symmetrical"
if kurt > 0:
    kurtosis_description = "leptokurtic (fat tails)"
elif kurt < 0:
    kurtosis_description = "platykurtic (thin tails)"
else:
    kurtosis_description = "mesokurtic (normal tails)"

print(f"The dataset is {skewness_description}.")
print(f"The dataset is {kurtosis_description}.")
```
